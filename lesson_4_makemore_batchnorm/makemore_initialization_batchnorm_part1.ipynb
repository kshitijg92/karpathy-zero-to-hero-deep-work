# Section 1: Initialization

We briefly touched upon uncharacteristically high losses during the first step of training with a randomly initialized model during the [previous lesson notebook](../lesson_3_makemore_mlp/makemore_mlp.ipynb#why-is-the-initial-loss-so-high?). We will now look into weight initialization further.

**Why initialization matters so much:** In a neural network, the initial weights determine the initial output distribution (predictions), the initial gradient magnitudes, and consequently the trajectory of the entire optimization. Poor initialization leads to:

- **Confidently wrong predictions** at step 0 → enormous initial loss → huge gradients that "waste" early training steps correcting the overconfident output
- **Saturated activations** (e.g. tanh outputs near ±1) → vanishing gradients → slow or stalled learning
- **Exploding activations** → numerical instability → NaN losses

The goal is to find an initialization where the network starts in a "reasonable" state: uniform predictions (not confidently wrong), activations in a useful range, and gradients that flow smoothly through every layer.


````python

import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
````

````python
# lets copy over some stuff from the previous notebook

# utility to read dataset
DATASET_PATH = '../names.txt'
SPECIAL_TOKEN = "."

def get_dataset():
    with open(DATASET_PATH, 'r') as f:
        rows = [row.strip() for row in f.readlines()]
    return rows

# Load dataset
words = get_dataset()
print(f"{len(words)} names loaded")
print(f"Examples: {words[:8]}")

# Build character mappings — identical to lesson 2
# '.' is our special start/end token at index 0, then a=1, b=2, ..., z=26
all_characters = [SPECIAL_TOKEN] + sorted(list(set(''.join(words))))
stoi = {s: i for i, s in enumerate(all_characters)}
itos = {i: s for s, i in stoi.items()}
vocab_size = len(itos)
print(f"Vocabulary size: {vocab_size}")
print(f"Mappings: {itos}")
````

````python
def build_dataset(words, block_size=3):
    """
    Convert a list of words into (X, Y) tensors for training.

    X shape: (N, block_size) — each row is a context window of character indices
    Y shape: (N,) — each element is the target character index

    This function will be called three times: once each for train, val, and test splits.
    """
    X, Y = [], []
    for w in words:
        context = [0] * block_size
        for ch in w + SPECIAL_TOKEN:
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    return X, Y

import random
random.seed(42)
random.shuffle(words)

# 80/10/10 split at the word level
n1 = int(0.8 * len(words))
n2 = int(0.9 * len(words))

Xtr, Ytr = build_dataset(words[:n1], block_size=3)     # training
Xdev, Ydev = build_dataset(words[n1:n2], block_size=3)  # validation ("dev" set)
Xte, Yte = build_dataset(words[n2:], block_size=3)      # test

print(f"Training:   {Xtr.shape[0]:>7,} examples from {n1:,} names")
print(f"Validation: {Xdev.shape[0]:>7,} examples from {n2-n1:,} names")
print(f"Test:       {Xte.shape[0]:>7,} examples from {len(words)-n2:,} names")
````

### Model architecture (carried over from Lesson 3)

We're using the same character-level MLP from the [previous notebook](../lesson_3_makemore_mlp/makemore_mlp.ipynb). The architecture is:

```
Input (3 chars) → Embedding (10-dim each) → Concat (30-dim) → Linear+tanh (200 hidden) → Linear (27 logits) → Softmax → Loss
```

In tensor terms:

- **`C`** `(27, 10)`: Embedding lookup table — each of the 27 characters gets a 10-dimensional vector
- **`W1`** `(30, 200)` + **`b1`** `(200,)`: Hidden layer weights. The input is the 3 embeddings concatenated (3 × 10 = 30), producing 200 hidden activations passed through `tanh`
- **`W2`** `(200, 27)` + **`b2`** `(27,)`: Output layer weights. Maps 200 hidden features → 27 logits (one per character)

Total parameters: 27×10 + 30×200 + 200 + 200×27 + 27 = **11,897**

The `train_model` function below runs the standard mini-batch SGD loop: sample 32 examples → forward pass → cross-entropy loss → backward pass → parameter update. It returns the final-step activations (`h`, `hidden_layer_preactivation`, `logits`) so we can inspect them for initialization issues.


````python
def get_params(embed_dim, block_size, n_hidden):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                        generator=g)
    W2 = torch.randn((n_hidden, vocab_size),           generator=g)
    b2 = torch.randn(vocab_size,                       generator=g)
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

@torch.no_grad()
def eval_loss(X, Y, params, embed_dim, block_size):
    C, W1, b1, W2, b2, parameters = params
    emb = C[X]
    h = torch.tanh(emb.view(-1, embed_dim * block_size) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)
    return loss.item()

def train_model(params, n_steps=200000, batch_size=32):

    C, W1, b1, W2, b2, parameters = params

    stepi = []
    lossi = []
    loglossi = []

    pbar = tqdm(range(n_steps), desc="Training")
    for i in pbar:
        # Mini-batch: randomly sample 32 examples
        ixds = torch.randint(0, Xtr.shape[0], (batch_size,))
        mini_batch_inp, mini_batch_target = Xtr[ixds], Ytr[ixds]

        # Forward pass
        emb = C[mini_batch_inp]                                        # (32, 3, 10)
        emb_cat = emb.view(emb.shape[0], -1)                            # (32, 30)
        hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)
        h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
        logits = h @ W2 + b2                                           # (32, 27)
        loss = F.cross_entropy(logits, mini_batch_target)

        # Backward pass
        for p in parameters:
            p.grad = None
        loss.backward()

        # Learning rate step decay: 0.1 for first 100K steps, then 0.01
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad

        # Track loss
        stepi.append(i)
        lossi.append(loss.item())
        loglossi.append(loss.log10().item())

        if i % 10 == 0:
            pbar.set_postfix(loss=f"{loss.data:.4f}")

    return stepi, lossi, loglossi, hidden_layer_preactivation, h, logits

def plot_loss(stepi, lossi):
    # Plot training loss over time
    plt.figure(figsize=(10, 4))
    plt.plot(stepi, lossi, alpha=0.15, color='tab:blue', label='Per-batch loss')

    # Moving average (window of 500 steps) to show the trend clearly
    window = 500
    moving_avg = torch.tensor(lossi).unfold(0, window, 1).mean(dim=1)
    plt.plot(stepi[window-1:], moving_avg, color='red', linewidth=2, label=f'Moving avg (window={window})')

    plt.xlabel('Step')
    plt.ylabel('Loss (log10)')
    plt.title('Training Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
````

````python
embed_dim = 10
block_size = 3
n_hidden = 200

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden)
stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters], n_steps=1)

print(f"Loss after a single step when model is randomly initialized: {lossi}")
````

````python
print(f"Expected random loss for {vocab_size=} is {-torch.log(torch.tensor(1./vocab_size)):.3f}")
````

Now lets think about this again for a second. If a model is untrained or random, then our expectation is that any single character can be equally likely to be predicted. For a vocab size of `27`, our loss for the first step of forward pass should be ~`3.29`. Why is our loss so high?

It turns out its because of the way our weights are initialized. For our output probability distribution to be uniform, the logits should all be equal to 0 (or any other identical constant number). This is because we obtain our probabilities through a softmax of the logits.

**The math behind this:** Recall that softmax converts logits $z_i$ into probabilities:

$$p_i = \frac{e^{z_i}}{\sum_j e^{z_j}}$$

If all logits are equal (say $z_i = c$ for all $i$), then every $e^{z_i}$ is the same, so every probability becomes $\frac{1}{N}$ — a uniform distribution. The cross-entropy loss for a uniform distribution over 27 classes is $-\log(\frac{1}{27}) \approx 3.296$.

But with randomly initialized weights, our logits span a huge range (as we'll see below). The softmax function is exponential, so even modest differences between logits get amplified into wildly skewed probabilities. The model ends up assigning nearly 100% probability to one class and nearly 0% to everything else — it's **confidently wrong**.


````python
# lets check what our logits look like after first step of training
logits[0]
````

We can see that our logits span a wide range of values (roughly -26 to +36). Thus the softmax fn pushes some character probability very high while pushing down all the others. In other words, our model is very confidently wrong. Which is undesirable. We would want our random model to be uniformly wrong.

**Why are the logits so extreme?** Let's trace through the computation. Our logits are computed as:

```python
logits = h @ W2 + b2
```

Each element of the logits vector is a dot product: $\text{logit}_j = \sum_{i=1}^{d} h_i \cdot W2_{ij} + b2_j$, where $d = 200$ (the hidden layer size). Since `W2` is initialized from $\mathcal{N}(0, 1)$ and `b2` is also from $\mathcal{N}(0, 1)$, each logit is a sum of 200 random products plus a random bias.

**The Central Limit Theorem (CLT) tells us what to expect here.** The CLT says that when you sum up a large number of independent random variables, the result approaches a Gaussian distribution regardless of the original distributions. More precisely, if $X_1, X_2, \ldots, X_n$ are i.i.d. with mean $\mu$ and variance $\sigma^2$, then:

$$\frac{1}{n}\sum_{i=1}^{n} X_i \xrightarrow{d} \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)$$

or equivalently, the _sum_ (not the mean) has variance $n\sigma^2$, so std $= \sigma\sqrt{n}$.

In our case, each term $h_i \cdot W2_{ij}$ is a product of two values. The $h_i$ values come from `tanh`, so they're bounded in $[-1, 1]$ with some variance (let's approximate it as ~1 for a rough calculation). Each $W2_{ij} \sim \mathcal{N}(0, 1)$. Each product $h_i \cdot W2_{ij}$ has mean 0 and variance roughly 1. Summing 200 such terms gives us a logit with:

- **Mean:** 0 (good — centered)
- **Std:** $\approx \sqrt{200} \approx 14.1$ (bad — way too spread out)

So our logits end up distributed roughly as $\mathcal{N}(0, 14^2)$, which explains why we see values in the ±30 range (within ~2 standard deviations). For softmax to produce a near-uniform distribution, we'd need the logits to have std close to 0 — they're off by a factor of ~14.

This is the core insight behind proper initialization: **we need to scale weights down to compensate for the fan-in** (the number of inputs to each neuron). The larger the fan-in, the more we need to shrink the weights to prevent the output from exploding. This same reasoning will lead us to Kaiming initialization later.


````python
# Lets explore this with a sample 4-dim tensor
sample_logits = torch.tensor([0., 0., 0., 0.])
probs = torch.softmax(sample_logits, dim=0)
loss = -probs[2].log() # assuming that "2" was the correct label
sample_logits, probs, loss
````

For this sample example, we can see that when the logits are all zero, we get a uniform probability distribution — each class gets exactly $\frac{1}{4} = 0.25$. The loss is $-\log(0.25) = \log(4) \approx 1.386$, which is exactly the "fair" loss for a 4-class problem.

What happens if these logits take very extreme values like the ones above from our random model? We should see a much higher loss and skewed probability distribution.


````python
# Lets explore this with a sample 4-dim tensor
sample_logits = torch.tensor([10., -10., 4., -3.])
probs = torch.softmax(sample_logits, dim=0)
loss = -probs[2].log() # assuming that "2" was the correct label
sample_logits, probs, loss
````

Indeed — with logits of `[10, -10, 4, -3]`, the softmax assigns 99.75% of the probability mass to the first class. The correct class (index 2) gets only 0.25%, so the loss shoots up to ~6.0 — nearly 4x worse than the uniform case. And our real model's logits are even more extreme than this toy example.

Okay, so how do we fix this? What we want is for our `logits` to be near `0` at initialization. If we take a look at the logits calculation, we see

```python
logits = h @ W2 + b2
```

Based on the calculation above, if we set both `W2` and `b2` to exactly `0`, then we would get initial logits to be `0`.


````python
embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                        generator=g)
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * scale_w2 # seto to 0
    b2 = torch.randn(vocab_size,                       generator=g) * scale_b2 # set to 0
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters], n_steps=1)

print(f"Loss after a single step when model is randomly initialized: {lossi}")
print(f"Since all logits are 0, {torch.sum(logits).item()=}")
````

We can see that when logits are `0` because of the way we initialize the weights, our initial loss is exactly what we expected from predicting a uniform distribution.

This is a good sign — it means the network isn't wasting its first training steps fighting an overconfident output distribution. But does this mean setting `W2 = 0` is a good initialization strategy? Let's find out by training for the full 200k steps.


````python
# lets now train this network for the entire 200k steps
embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                        generator=g)
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * scale_w2 # seto to 0
    b2 = torch.randn(vocab_size,                       generator=g) * scale_b2 # set to 0
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters])
print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}")
````

````python
print(f"Final training loss: {lossi[-1]}")
plot_loss(stepi, lossi)
````

We can see that we are actually not doing much better than before, in fact we are doing **worse**. (For same model, we had training loss of `2.3501` in our previous notebook).

This is because we initialized the weights `W2` to 0, introducing a symmetry and making it much harder for the model to pass gradients through during backprop.

### The symmetry problem with W2 = 0

To understand why zero initialization is so damaging, let's trace exactly what happens during the **first forward and backward pass** when `W2 = 0`.

**Forward pass:**

```
logits = h @ W2 + b2
```

Since `W2 = 0` and `b2 = 0`, all logits are exactly 0 → softmax gives a uniform distribution → loss ≈ 3.296. So far so good — this is the "correct" initial loss.

**Backward pass — the problem emerges:**

The gradient flows backward through the network. Let's follow it:

1. **$\frac{\partial \mathcal{L}}{\partial \text{logits}}$**: This is `softmax(logits) - one_hot(target)`. Since softmax is uniform ($\frac{1}{27}$ for each class), this gradient is a vector that's roughly $\frac{1}{27} \approx 0.037$ everywhere, with $\frac{1}{27} - 1 \approx -0.963$ at the correct class index. This is fine — nonzero gradients.

2. **$\frac{\partial \mathcal{L}}{\partial W_2} = h^T \cdot \frac{\partial \mathcal{L}}{\partial \text{logits}}$**: Since $h$ (the tanh activations) is nonzero and $\frac{\partial \mathcal{L}}{\partial \text{logits}}$ is nonzero, this gradient is nonzero. So `W2` **does** get updated — it will move away from zero after one step. ✓

3. **$\frac{\partial \mathcal{L}}{\partial h} = \frac{\partial \mathcal{L}}{\partial \text{logits}} \cdot W_2^T$**: Here's the critical problem: **$W_2$ is still zero during this backward pass** (parameters are updated _after_ the full backward pass, not during it). So $\frac{\partial \mathcal{L}}{\partial h} = \frac{\partial \mathcal{L}}{\partial \text{logits}} \cdot \mathbf{0}^T = \mathbf{0}$. The gradient signal **dies completely** at this point.

4. **$\frac{\partial \mathcal{L}}{\partial W_1}$**: This depends on $\frac{\partial \mathcal{L}}{\partial h}$ flowing backward through the tanh. But $\frac{\partial \mathcal{L}}{\partial h} = 0$, so $\frac{\partial \mathcal{L}}{\partial W_1} = 0$. The first layer receives **no learning signal whatsoever**.

The result: after the first step, `W2` gets a small update but `W1` stays exactly where it started. On the second step, `W2` is now slightly nonzero, so a tiny gradient trickles back to `W1` — but it's very small. It takes many steps for `W1` to start receiving meaningful gradients, and during all those steps, the network is effectively only learning through its output layer. The hidden layer — which is responsible for learning useful features — is essentially frozen at its random initialization for a long time.

Let's verify this with a tiny example below.


````python
# Worked example: tiny network (2 inputs → 10 hidden → 27 outputs)
# Running for 5 steps to see how gradients evolve over time
# Using a larger output dim and small lr so the effect is clearly visible

torch.manual_seed(0)

x = torch.randn(1, 2)
target = torch.tensor([5])      # correct class is 5 (out of 27)
lr = 0.001
n_hidden_demo = 10
n_out_demo = 27

print("=" * 60)
print("CASE 1: W2 initialized to ZERO")
print("=" * 60)

W1_c1 = torch.randn(2, n_hidden_demo, requires_grad=True)
W2_c1 = torch.zeros(n_hidden_demo, n_out_demo, requires_grad=True)
b2_c1 = torch.zeros(n_out_demo, requires_grad=True)

for step in range(5):
    h = torch.tanh(x @ W1_c1)
    logits = h @ W2_c1 + b2_c1
    loss = F.cross_entropy(logits, target)
    loss.backward()

    w1_grad_norm = W1_c1.grad.norm().item()
    w2_grad_norm = W2_c1.grad.norm().item()
    print(f"  Step {step}: loss={loss.item():.4f}  "
          f"||∂L/∂W1||={w1_grad_norm:.6f}  "
          f"||∂L/∂W2||={w2_grad_norm:.6f}")

    with torch.no_grad():
        W1_c1 -= lr * W1_c1.grad
        W2_c1 -= lr * W2_c1.grad
        b2_c1 -= lr * b2_c1.grad
    W1_c1.grad = None
    W2_c1.grad = None
    b2_c1.grad = None

print(f"\n{'=' * 60}")
print("CASE 2: W2 initialized to SMALL values (0.01)")
print("=" * 60)

torch.manual_seed(0)  # reset so W1 starts the same

W1_c2 = torch.randn(2, n_hidden_demo, requires_grad=True)
W2_c2 = (torch.randn(n_hidden_demo, n_out_demo) * 0.01).requires_grad_(True)
b2_c2 = torch.zeros(n_out_demo, requires_grad=True)

for step in range(5):

    h = torch.tanh(x @ W1_c2)
    logits = h @ W2_c2 + b2_c2
    loss = F.cross_entropy(logits, target)
    loss.backward()

    w1_grad_norm = W1_c2.grad.norm().item()
    w2_grad_norm = W2_c2.grad.norm().item()
    print(f"  Step {step}: loss={loss.item():.4f}  "
          f"||∂L/∂W1||={w1_grad_norm:.6f}  "
          f"||∂L/∂W2||={w2_grad_norm:.6f}")

    with torch.no_grad():
        W1_c2 -= lr * W1_c2.grad
        W2_c2 -= lr * W2_c2.grad
        b2_c2 -= lr * b2_c2.grad
    W1_c2.grad = None
    W2_c2.grad = None
    b2_c2.grad = None
````

The example confirms exactly what we predicted:

- **Case 1 ($W_2 = 0$):** At step 0, $\left\|\frac{\partial \mathcal{L}}{\partial W_1}\right\|$ is exactly 0 — the hidden layer is completely frozen. After step 0 updates $W_2$ to something slightly nonzero, a tiny gradient starts trickling back to $W_1$ in subsequent steps, but it's orders of magnitude smaller than in Case 2. The hidden layer is learning, but _agonizingly slowly_.
- **Case 2 ($W_2 = 0.01 \times \text{random}$):** $\left\|\frac{\partial \mathcal{L}}{\partial W_1}\right\|$ is nonzero from step 0. Both layers learn immediately.

The key formula to remember: $\frac{\partial \mathcal{L}}{\partial h} = \frac{\partial \mathcal{L}}{\partial \text{logits}} \cdot W_2^T$. If $W_2 = 0$, then $W_2^T = 0$, and the entire backward signal is multiplied by zero — killing all gradients to earlier layers.

So, instead of setting `W2` to `0`, lets scale it by `0.01` — small enough that logits start near zero, but nonzero so gradients can flow.


````python
# lets now train this network for the entire 200k steps
embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                        generator=g)
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * scale_w2 # seto to 0
    b2 = torch.randn(vocab_size,                       generator=g) * scale_b2 # set to 0
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden, scale_w2 = 0.01, scale_b2 = 0)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters])
print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}")
````

````python
plot_loss(stepi, lossi)
````

We can see that by properly initializing the weight `W2` and the bias `b2`, we are getting much better results with training loss of `1.9`. Another thing to note here is that our loss is no longer extremely hockey-stick shaped like before. This is because in the badly initialized network, the model spent the first few thousand steps just trying to squish the weights, as opposed to doing any actual learning. When we properly initialized the network, the model was already starting to learn within the initial steps.

**The hockey-stick problem:** With random initialization, the initial loss was ~24 (vs the expected ~3.3). Those first thousands of steps aren't really "learning" the structure of names — they're just fighting to bring the output distribution from wildly overconfident back down to something reasonable. This is wasted optimization budget. By initializing `W2` small, we skip that entire phase and the network immediately begins learning useful patterns.

**A general principle:** For the final (output) layer of a classification network, it's common practice to initialize the weights very small (or zero for biases) so that the initial predictions are approximately uniform. PyTorch's default `nn.Linear` initialization already does something similar — it draws from $\mathcal{U}(-\frac{1}{\sqrt{d}}, \frac{1}{\sqrt{d}})$ where $d$ is the fan-in, which keeps initial outputs small.

Are we done though? _Not really_. We've fixed the output layer, but we haven't looked at what's happening _inside_ the network. Next lets take a look at the activations after the non-linearity from our first linear layer — the `tanh` outputs. If those are problematic, we'll have another issue to address.


````python
# lets now train this network for the entire 200k steps
embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden, scale_w2 = 0, scale_b2 = 0):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                        generator=g)
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * scale_w2
    b2 = torch.randn(vocab_size,                       generator=g) * scale_b2
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden, scale_w2 = 0.01, scale_b2 = 0)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters], n_steps=1)
print(f"Final training loss: {lossi[-1]}")
````

## Section 1.2: Hidden layer activations — the tanh saturation problem

We've fixed the output layer so the initial loss is reasonable. But what about the _inside_ of the network? Let's inspect the activations coming out of the hidden layer's `tanh` nonlinearity:

```python
hidden_layer_preactivation = emb_cat @ W1 + b1
h = torch.tanh(hidden_layer_preactivation)
```

Remember, `tanh` squashes its input to the range $[-1, 1]$. If the pre-activation values going _into_ `tanh` are very large (positive or negative), the outputs will be pushed to the extreme ends of this range — we call this **saturation**. Let's see if that's happening.

### Why tanh saturation kills gradients

The histogram on the left shows that almost all activations are pinned at exactly -1 or +1. The histogram on the right reveals why: the pre-activation values (`emb_cat @ W1 + b1`) span a huge range (roughly -15 to +15), so `tanh` is operating deep in its flat tails.

Why is this a problem? Recall the backward pass through `tanh` from our [micrograd notebook](../lesson_1_micrograd/micrograd.ipynb):

```python
def tanh(self):
    x = self.data
    tanh_val = (math.exp(2*x) - 1) / (math.exp(2*x) + 1)
    out = Value(tanh_val, (self, ), 'tanh')

    def backward():
        self.grad += (1 - tanh_val**2) * out.grad

    out._backward = backward
    return out
```

The local gradient of `tanh` is $(1 - \tanh^2(x))$. Consider two cases:

1. **`tanh(x) ≈ 0`** (pre-activation near 0): The local gradient is $(1 - 0^2) = 1$. The `tanh` node acts as a **pass-through** — gradients flow through undiminished.
2. **`tanh(x) ≈ ±1`** (pre-activation far from 0 — our situation): The local gradient is $(1 - 1^2) = 0$. The `tanh` node becomes a **gradient killer** — it multiplies the incoming gradient by ~0, blocking all signal from flowing backward.

This is the **vanishing gradient problem** in action. When neurons are saturated, they become **dead neurons** — they can't learn because no gradient flows through them. This makes intuitive sense too: `tanh` is a squashing function, so for any very large input, small changes to that input still get squashed to the same ±1 output, meaning the derivative is near zero.

Note that this problem isn't unique to `tanh` — **sigmoid** has the same flat-tail issue, and **ReLU** has a related problem where negative inputs produce exactly zero gradient (the "dying ReLU" problem).

In our case, we were still able to train the model because `tanh` doesn't saturate for _all_ training examples. For some examples, the pre-activations happen to be near zero, allowing gradients to pass through — but this is wasteful. We can see this in the next cell.


````python
print(h.shape)
plt.figure(figsize=(20, 10))
plt.imshow(h.abs() > 0.99, cmap='gray', interpolation='nearest');
````

In the above plot, the white cells denote where tanh is saturated (not allowing gradients), while black shows where it isn't. The rows are the 32 examples in our mini-batch; the columns are the 200 neurons. We can see that no column is entirely white, which means every neuron is activated (non-saturated) for _at least some_ inputs — so no neuron is truly "dead." But the vast majority of cells are white, meaning most neurons are saturated for most inputs, which is still very wasteful.

**Dead neurons during training:** Even with perfect initialization, dead neurons can appear _during_ training. If the learning rate is too high, a large gradient update can push a neuron's weights so far that it falls off the data manifold — from that point on, no input example will ever activate it, and it's permanently dead. This is one reason why learning rate tuning matters.

### How do we fix this?

The same way we fixed the logits — by controlling the scale of the pre-activations.

```python

h = torch.tanh(hidden_layer_preactivation)
```

If we want `h` to take on a range of values (not just ±1), we need `hidden_layer_preactivation` to be near 0 at initialization — in the linear region of `tanh` where it hasn't saturated yet. We can achieve this by scaling down `W1` and `b1`, just like we did for `W2` and `b2` with the logits.


````python
embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden, scale_w1 = 0, scale_b1 = 0, scale_w2 = 0, scale_b2 = 0):
    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g) * scale_w1
    b1 = torch.randn(n_hidden,                        generator=g) * scale_b1
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * scale_w2
    b2 = torch.randn(vocab_size,                       generator=g) * scale_b2
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden, scale_w1 = 0.2, scale_b1 = 0.01, scale_w2 = 0.01, scale_b2 = 0)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters], n_steps=1)
print(f"Final training loss: {lossi[-1]}")
````

### The problem with hand-tuned initialization

This illustrates the impact of initialization on training — by carefully choosing scale factors for `W1`, `b1`, `W2`, and `b2`, we got the initial loss right, avoided tanh saturation, and achieved reasonable training loss.

But we picked those scale factors (`0.2`, `0.01`, etc.) **by hand**, through trial and error. This approach doesn't scale:

- Different architectures need different scales
- Adding or removing layers changes what works
- Changing the hidden size, embedding size, or any other hyperparameter invalidates the previous tuning

We need a **principled, automatic** way to choose initialization scales based on the network architecture. This is exactly what **Kaiming (He) initialization** provides — a formula derived from the statistics of the forward pass that tells us the right scale for each layer based on its fan-in. We'll cover that next.

## Section 1.3: Deriving the correct weight scale

Let's motivate the correct scaling with a simple experiment. We'll take a Gaussian input and multiply it by a weight matrix, ignoring bias and nonlinearity for now, and see what happens to the output distribution.


````python
# Simplified experiment: what happens to the distribution when we multiply by a weight matrix?
# No bias, no nonlinearity — just y = x @ w
x = torch.randn(1000, 30)   # 1000 samples, 30 features (fan_in = 30)
w = torch.randn(30, 200)    # weights from N(0, 1) — no scaling
y = x @ w

print(f"Input X, {x.mean()=}, {x.std()=}")
print(f"Output Y, {y.mean()=}, {y.std()=}")

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].hist(x.flatten().tolist(), bins=50, density=True)
axs[0].set_title('Input X distribution')

axs[1].hist(y.flatten().tolist(), bins=50, density=True)
axs[1].set_title('Output Y distribution')

plt.tight_layout()
plt.show()
````

We can see that if we have an input Gaussian $X$ with mean 0 and unit std, the output $Y = X \cdot W$ has a much higher std (~5.4). The output distribution is much more spread out, which is exactly the problem we've been fighting. In a deep network, this effect **compounds layer after layer** — if each layer amplifies the std by ~5x, then after just 3 layers the activations would have std $\approx 5^3 = 125$, completely blowing up.

We want the distributions throughout every layer of the neural network to be approximately Gaussian with mean 0 and std ~1. This is the **variance preservation** principle: each layer should neither amplify nor shrink the scale of its inputs. How can we achieve that?

### Deriving the correct scale from first principles

Consider a single output element $y_j$ of the matrix multiply $Y = X \cdot W$:

$$y_j = \sum_{i=1}^{n_\text{in}} x_i \cdot W_{ij}$$

where $n_\text{in}$ is the **fan-in** — the number of input features (here 30). Each $x_i$ and $W_{ij}$ are independent random variables. Let's compute the variance of $y_j$:

**Step 1 — Variance of a single product:** If $x_i$ and $W_{ij}$ are independent with mean 0, then:

$$\text{Var}(x_i \cdot W_{ij}) = \text{Var}(x_i) \cdot \text{Var}(W_{ij})$$

(This holds because for zero-mean independent variables, $E[(xw)^2] = E[x^2]E[w^2] = \text{Var}(x)\text{Var}(w)$.)

With $x_i \sim \mathcal{N}(0, 1)$ and $W_{ij} \sim \mathcal{N}(0, \sigma_w^2)$:

$$\text{Var}(x_i \cdot W_{ij}) = 1 \cdot \sigma_w^2 = \sigma_w^2$$

**Step 2 — Variance of the sum:** Since all $n_\text{in}$ terms are independent, variances add:

$$\text{Var}(y_j) = \sum_{i=1}^{n_\text{in}} \text{Var}(x_i \cdot W_{ij}) = n_\text{in} \cdot \sigma_w^2$$

**Step 3 — Solve for variance preservation:** We want $\text{Var}(y_j) = \text{Var}(x_i) = 1$:

$$n_\text{in} \cdot \sigma_w^2 = 1 \quad \Rightarrow \quad \sigma_w = \frac{1}{\sqrt{n_\text{in}}}$$

So if we initialize $W_{ij} \sim \mathcal{N}\left(0, \frac{1}{n_\text{in}}\right)$ — i.e., initialize our random weights as $W_{ij} \sim \mathcal{N}\left(0, 1\right)$ and then scale by $\frac{1}{\sqrt{n_\text{in}}}$ — the output will have the same std as the input. In our example with fan-in of 30, this gives $\sigma_w = \frac{1}{\sqrt{30}} \approx 0.183$.

Let's verify this by trying different weight scales below.


````python
# Experiment 1: weights too LARGE — activations explode
weight_scale = 10
x = torch.randn(1000, 30)
w = torch.randn(30, 200) * weight_scale
y = x @ w

print(f"Scaling weights by {weight_scale} ")
print(f"Input X, {x.mean()=}, {x.std()=}")
print(f"Output Y, {y.mean()=}, {y.std()=}")

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].hist(x.flatten().tolist(), bins=50, density=True)
axs[0].set_title('Input X distribution')

axs[1].hist(y.flatten().tolist(), bins=50, density=True)
axs[1].set_title('Output Y distribution')

plt.tight_layout()
plt.show()

print ("-"*20)

# Experiment 2: weights too SMALL — activations shrink toward zero
weight_scale = 0.04
x = torch.randn(1000, 30)
w = torch.randn(30, 200) * weight_scale
y = x @ w

print(f"Scaling weights by {weight_scale} ")
print(f"Input X, {x.mean()=}, {x.std()=}")
print(f"Output Y, {y.mean()=}, {y.std()=}")

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].hist(x.flatten().tolist(), bins=50, density=True)
axs[0].set_title('Input X distribution')

axs[1].hist(y.flatten().tolist(), bins=50, density=True)
axs[1].set_title('Output Y distribution')

plt.tight_layout()
plt.show()

print ("-"*20)

# Experiment 3: weights scaled by 1/√fan_in — variance is preserved!
weight_scale = 1.0 / 30 ** 0.5
x = torch.randn(1000, 30)
w = torch.randn(30, 200) * weight_scale
y = x @ w

print(f"Scaling weights by {weight_scale} ")
print(f"Input X, {x.mean()=}, {x.std()=}")
print(f"Output Y, {y.mean()=}, {y.std()=}")

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].hist(x.flatten().tolist(), bins=50, density=True)
axs[0].set_title('Input X distribution')

axs[1].hist(y.flatten().tolist(), bins=50, density=True)
axs[1].set_title('Output Y distribution')

plt.tight_layout()
plt.show()
````

The three experiments confirm the theory:

| Weight scale                 | Output std | Effect                                                                        |
| ---------------------------- | ---------- | ----------------------------------------------------------------------------- |
| `10` (too large)             | ~55        | Activations explode — would cause saturation and numerical issues             |
| `0.04` (too small)           | ~0.2       | Activations shrink toward zero — network has very weak signals, learns slowly |
| `1/√30 ≈ 0.183` (just right) | ~1.0       | Output std matches input std — **variance is preserved**                      |

The correct scaling factor is $\frac{1}{\sqrt{n_\text{in}}}$ where $n_\text{in}$ is the **fan-in** (number of inputs to each neuron). This is called **Xavier/Glorot initialization** (2010), and it works well for activations like `sigmoid` and `tanh` that are roughly linear near zero.

### What about the nonlinearity? The "gain" factor

Our derivation above assumed a purely linear layer ($y = x \cdot W$) with no activation function. But in practice, the output of each layer passes through a nonlinearity like `tanh` or `ReLU`, which _changes_ the variance of the signal. The **gain** is a correction factor that accounts for this.

**For `tanh`:** Near zero, `tanh(x) ≈ x` — it's approximately linear, so it roughly preserves variance. But for larger inputs, `tanh` squashes values toward ±1, which slightly _reduces_ variance. The recommended gain for `tanh` is $\frac{5}{3} \approx 1.667$, meaning we should slightly _inflate_ the weights to compensate for this squashing:

$$\sigma_w = \frac{5}{3} \cdot \frac{1}{\sqrt{n_\text{in}}}$$

**For `ReLU`:** ReLU($x$) = max(0, $x$) zeros out all negative values, which cuts roughly half the signal. This halves the variance at each layer. To compensate, we need to double the variance of the weights, giving a gain of $\sqrt{2}$:

$$\sigma_w = \sqrt{2} \cdot \frac{1}{\sqrt{n_\text{in}}} = \sqrt{\frac{2}{n_\text{in}}}$$

This is the **Kaiming/He initialization** (He et al., 2015), which is the default in modern deep learning frameworks for ReLU networks.

**For a linear layer (no activation):** The gain is simply 1 — no correction needed.

In PyTorch, you can access these gains via `torch.nn.init.calculate_gain('tanh')` (returns 5/3) and `torch.nn.init.calculate_gain('relu')` (returns √2). The general Kaiming formula is:

$$W \sim \mathcal{N}\left(0, \frac{\text{gain}^2}{n_\text{in}}\right)$$

The key insight is the same across all variants: **scale your weights by the inverse square root of the fan-in**, adjusted by a gain factor that depends on the activation function, to preserve the distribution of activations across layers.


````python
# Lets now train our network with this automated initialization of weights

embed_dim = 10
block_size = 3
n_hidden = 200

def get_params(embed_dim, block_size, n_hidden):

    tanh_gain = 5./3  # correction factor for tanh squashing variance

    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g) * (tanh_gain / (embed_dim * block_size) ** 0.5)  # Kaiming with tanh gain
    b1 = torch.randn(n_hidden,                        generator=g) * 0.01  # small, near zero
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * (1 / (n_hidden) ** 0.5)  # Xavier (gain=1, no activation)
    b2 = torch.randn(vocab_size,                       generator=g) * 0.01  # small, near zero
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters])
print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}")
````

### Putting it all together: Kaiming initialization for our model

Now let's apply the principled initialization we just derived to our actual network, replacing the hand-tuned scale factors with the correct formulas:

- **`W1`** (hidden layer with `tanh`): Scale by $\frac{\text{gain}}{\sqrt{n_\text{in}}} = \frac{5/3}{\sqrt{30}} \approx 0.304$. The gain of $\frac{5}{3}$ compensates for `tanh` slightly reducing variance. Note that this is larger than our hand-tuned `0.2` — we were actually under-scaling the hidden layer before.
- **`W2`** (output layer, no activation): Scale by $\frac{1}{\sqrt{n_\text{in}}} = \frac{1}{\sqrt{200}} \approx 0.071$. No gain correction needed since the output layer has no nonlinearity — we just want the logits to start small.
- **`b1`, `b2`** (biases): Initialize very small (`0.01`). Biases don't participate in the fan-in scaling since they're added after the matrix multiply. Setting them near zero is standard practice — they'll learn the right offsets during training.

This is a significant improvement over our earlier approach: instead of 4 arbitrary hyperparameters (`scale_w1=0.2`, `scale_b1=0.01`, `scale_w2=0.01`, `scale_b2=0`), we now have a formula that automatically adapts to the layer dimensions. If we changed `n_hidden` from 200 to 500 or `embed_dim` from 10 to 20, the initialization would adjust accordingly.

### Results

The Kaiming-initialized model achieves comparable performance to our hand-tuned version, with the key advantage being that we didn't need to guess any scale factors — the formula computed them automatically from the layer dimensions. The loss curve shows no hockey-stick shape, the activations are well-distributed, and the saturation map should be mostly black.

**How this relates to PyTorch's defaults:** When you use `torch.nn.Linear(30, 200)` in PyTorch, it automatically initializes the weights using Kaiming uniform initialization: $W \sim \mathcal{U}\left(-\frac{1}{\sqrt{n_\text{in}}}, \frac{1}{\sqrt{n_\text{in}}}\right)$. The variance of a uniform distribution on $[-a, a]$ is $\frac{a^2}{3}$, so this gives $\text{Var}(W) = \frac{1}{3 \cdot n_\text{in}}$, which is slightly different from the Gaussian Kaiming init but achieves the same goal. You can also explicitly use `torch.nn.init.kaiming_normal_(W1, nonlinearity='tanh')` to get the exact initialization we derived here.


## Summary of Section 1: What we learned about initialization

This section traced a path from naive random initialization to principled Kaiming/He initialization. Here's the progression:

| Initialization         | Initial loss    | Training loss  | Problem                                                           |
| ---------------------- | --------------- | -------------- | ----------------------------------------------------------------- |
| `randn()` (no scaling) | ~24             | ~2.35          | Logits explode → confidently wrong predictions, hockey-stick loss |
| `W2 = 0, b2 = 0`       | 3.296 (perfect) | ~2.44 (worse!) | Zero weights block gradients to `W1` → hidden layer frozen        |
| `W2 *= 0.01, b2 = 0`   | ~3.3            | ~1.9           | **Hidden layer still saturated** — tanh pinned at ±1              |
| Hand-tuned scales      | ~3.3            | ~2.02          | Works, but fragile — scales are architecture-specific             |
| **Kaiming init**       | ~3.3            | ~2.0           | Principled, automatic, adapts to any layer size                   |

**The two key problems** we identified and solved:

1. **Output layer:** Logits must start near zero for uniform initial predictions. Scale `W2` by $\frac{1}{\sqrt{n_\text{hidden}}}$.
2. **Hidden layer:** Pre-activations must stay in the linear region of `tanh` to avoid saturation. Scale `W1` by $\frac{\text{gain}}{\sqrt{n_\text{in}}}$ where gain = $\frac{5}{3}$ for `tanh`.

**The unifying principle:** Each layer's weights should be scaled by $\frac{1}{\sqrt{n_\text{in}}}$ (times a gain factor for the nonlinearity) to preserve the variance of activations across layers. This prevents both explosion and vanishing.

**But there's a catch:** Kaiming init only guarantees good distributions at step 0. As training progresses, the distributions can drift — and as we saw, ~23% of neurons ended up saturated after 200K steps of training. Can we do better? That's where Batch Normalization comes in.


# Section 2: Batch Normalization

Batch Normalization was introduced in the 2015 paper [_Batch Normalization: Accelerating Deep Network Training by Reducing Internal Covariate Shift_](https://arxiv.org/abs/1502.03153) by Ioffe and Szegedy. It is one of the most influential techniques in deep learning — practically every modern architecture uses some form of normalization (BatchNorm, LayerNorm, GroupNorm, RMSNorm, etc.).


### The motivation: initialization is not enough

While proper initialization (Kaiming/He) ensured that activations followed a roughly zero-mean, unit-variance distribution **at step 0**, during training these activations are completely free to drift. As we saw from the saturation plots above, after 200K steps of training with Kaiming init, ~23% of the neurons' activations post-tanh became saturated — the distributions had shifted far from where they started.

The original paper called this phenomenon **internal covariate shift**: as the parameters of earlier layers change during training, the distribution of inputs to later layers shifts, forcing those layers to constantly re-adapt. (Note: the "internal covariate shift" explanation is somewhat [controversial in the research community](https://arxiv.org/abs/1805.11604) — BatchNorm clearly works, but the _reason_ it works may have more to do with smoothing the loss landscape than reducing covariate shift. Regardless, the practical technique is invaluable.)

The insight from the BatchNorm paper was beautifully simple: if you want your hidden states to be Gaussian with zero mean and unit std, then why not **normalize them to _be_ Gaussian**? Instead of hoping initialization keeps things well-behaved, just enforce it at every forward pass.


````python
# Lets copy over some functions

def plot_saturations(h, hidden_layer_preactivation):
    fig, axs = plt.subplots(2, 2, figsize=(14, 8),
                            gridspec_kw={'height_ratios': [1, 1.2]})

    # Post-tanh activations
    axs[0, 0].hist(h.flatten().tolist(), bins=50)
    axs[0, 0].set_title('Activations post tanh')

    # Pre-activation distribution
    axs[0, 1].hist(hidden_layer_preactivation.flatten().tolist(), bins=50)
    axs[0, 1].set_title('Pre-activations (before tanh)')

    # Saturation map — spans the full bottom row for a wider view
    ax_sat = plt.subplot(2, 1, 2)
    ax_sat.imshow(h.abs() > 0.99, cmap='gray', interpolation='nearest', aspect='auto')
    ax_sat.set_title('Saturation map (white = saturated)')
    ax_sat.set_xlabel('Neuron')
    ax_sat.set_ylabel('Example')

    plt.tight_layout()
    plt.show()

    print(f"Fraction of saturated neurons: {(h.abs() > 0.99).float().mean():.3f}")
    print(f"Pre-activation std: {hidden_layer_preactivation.std():.3f}")

def get_params(embed_dim, block_size, n_hidden):

    tanh_gain = 5./3  # correction factor for tanh squashing variance

    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g) * (tanh_gain / (embed_dim * block_size) ** 0.5)  # Kaiming with tanh gain
    b1 = torch.randn(n_hidden,                        generator=g) * 0.01  # small, near zero
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * (1 / (n_hidden) ** 0.5)  # Xavier (gain=1, no activation)
    b2 = torch.randn(vocab_size,                       generator=g) * 0.01  # small, near zero
    parameters = [C, W1, b1, W2, b2]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, parameters

def train_model(params, n_steps=200000, batch_size=32):

    C, W1, b1, W2, b2, parameters = params

    stepi = []
    lossi = []
    loglossi = []

    pbar = tqdm(range(n_steps), desc="Training")
    for i in pbar:
        # Mini-batch: randomly sample 32 examples
        ixds = torch.randint(0, Xtr.shape[0], (batch_size,))
        mini_batch_inp, mini_batch_target = Xtr[ixds], Ytr[ixds]

        # Forward pass
        emb = C[mini_batch_inp]                                        # (32, 3, 10)
        emb_cat = emb.view(emb.shape[0], -1)                            # (32, 30)
        hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)
        h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
        logits = h @ W2 + b2                                           # (32, 27)
        loss = F.cross_entropy(logits, mini_batch_target)

        # Backward pass
        for p in parameters:
            p.grad = None
        loss.backward()

        # Learning rate step decay: 0.1 for first 100K steps, then 0.01
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad

        # Track loss
        stepi.append(i)
        lossi.append(loss.item())
        loglossi.append(loss.log10().item())

        if i % 10 == 0:
            pbar.set_postfix(loss=f"{loss.data:.4f}")

    return stepi, lossi, loglossi, hidden_layer_preactivation, h, logits
````

### How BatchNorm works: normalizing the pre-activations

Let's look at our hidden layer's forward pass:

```python
hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)
h = torch.tanh(hidden_layer_preactivation)                       # (32, 200)
```

We want the activations flowing into `tanh` to be roughly unit Gaussian. This is the sweet spot for `tanh` — if the pre-activations are too small, `tanh` is approximately linear and the nonlinearity does nothing useful; if they're too large, `tanh` saturates and gradients vanish.

The BatchNorm operation normalizes each neuron's pre-activations **across the mini-batch**:

$$\hat{x}_j = \frac{x_j - \mu_j}{\sigma_j}$$

where $\mu_j$ and $\sigma_j$ are the mean and standard deviation of neuron $j$'s pre-activation computed over all examples in the current mini-batch. In code:

```python
_mean = hidden_layer_preactivation.mean(axis=0, keepdim=True)  # (1, 200) — one mean per neuron
_std  = hidden_layer_preactivation.std(axis=0, keepdim=True)   # (1, 200) — one std per neuron
hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std
```

Crucially, both `mean()` and `std()` are just arithmetic operations over tensors — they are fully differentiable. This means we can insert this normalization step into the computational graph and backpropagation will flow through it automatically. The gradients will tell the earlier layers how to adjust their weights in a way that accounts for the normalization.


````python
# Lets train the network by doing this now -
C, W1, b1, W2, b2, parameters = get_params(embed_dim, block_size, n_hidden)

def train_model(params, n_steps=200000, batch_size=32):

    C, W1, b1, W2, b2, parameters = params

    stepi = []
    lossi = []
    loglossi = []

    pbar = tqdm(range(n_steps), desc="Training")
    for i in pbar:
        # Mini-batch: randomly sample 32 examples
        ixds = torch.randint(0, Xtr.shape[0], (batch_size,))
        mini_batch_inp, mini_batch_target = Xtr[ixds], Ytr[ixds]

        # Forward pass
        emb = C[mini_batch_inp]                                        # (32, 3, 10)
        emb_cat = emb.view(emb.shape[0], -1)                            # (32, 30)
        hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)

        # CONVERT hidden_layer_preactivation to unit gaussian
        _mean = hidden_layer_preactivation.mean(axis=0, keepdim=True) # (1, 200) take mean across the samples in the mini batch
        _std = hidden_layer_preactivation.std(axis=0, keepdim=True) # (1, 200) take std across the samples in the mini batch

        hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std # convert to unit gaussian

        h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
        logits = h @ W2 + b2                                           # (32, 27)
        loss = F.cross_entropy(logits, mini_batch_target)

        # Backward pass
        for p in parameters:
            p.grad = None
        loss.backward()

        # Learning rate step decay: 0.1 for first 100K steps, then 0.01
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad

        # Track loss
        stepi.append(i)
        lossi.append(loss.item())
        loglossi.append(loss.log10().item())

        if i % 10 == 0:
            pbar.set_postfix(loss=f"{loss.data:.4f}")

    return stepi, lossi, loglossi, hidden_layer_preactivation, h, logits

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, parameters])
print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, parameters], embed_dim, block_size)}")
````

### The expressivity problem: raw normalization hurts performance

Two very important observations from the results above:

1. This network with BatchNorm is giving us **much worse results** than before (Before: `Training loss: 2.04, Validation Loss: 2.10` → Now: `Training loss: 2.77, Validation Loss: 2.82`)
2. We have **less than 1% saturated neurons** compared to 23% from before

So what happened? We fixed the activations to be Gaussian with zero mean and unit variance, but it came at a steep cost. We essentially **over-constrained** the network — by forcing every neuron's pre-activations to be a strict unit Gaussian at every step, we removed the network's ability to learn useful distribution shapes.

During training, the network needs freedom to:

- Make some distributions more diffuse or sharper
- Shift distributions away from zero (giving some neurons a bias toward firing or not firing)
- Make some `tanh` neurons more "trigger-happy" (saturated on purpose for specific inputs)

All of these adjustments are guided by backpropagation. By rigidly normalizing, we stripped the network of this expressivity — the hidden layer became a bottleneck.

### The fix: learnable scale and shift ($\gamma$ and $\beta$)

The BatchNorm paper's solution is elegant: after normalizing to unit Gaussian, apply a **learnable affine transformation**:

$$y_j = \gamma_j \cdot \hat{x}_j + \beta_j$$

where $\gamma_j$ (called `bngain` in our code) and $\beta_j$ (called `bnbias`) are per-neuron learnable parameters. At initialization, $\gamma = 1$ and $\beta = 0$, so the output is exactly the normalized value — a unit Gaussian. But during training, the network can learn to undo the normalization if that's what's optimal. In the extreme case, if the network learns $\gamma_j = \sigma_j$ and $\beta_j = \mu_j$, the BatchNorm becomes a no-op.

The key insight is that this gives the network the **best of both worlds**: it starts with well-behaved distributions (thanks to the initialization of $\gamma = 1, \beta = 0$), but can gradually learn to deviate from unit Gaussian as training progresses.


````python
# Lets add these scale and shift parameters

def get_params(embed_dim, block_size, n_hidden):

    tanh_gain = 5./3  # correction factor for tanh squashing variance

    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g) * (tanh_gain / (embed_dim * block_size) ** 0.5)  # Kaiming with tanh gain
    b1 = torch.randn(n_hidden,                        generator=g) * 0.01  # small, near zero
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * (1 / (n_hidden) ** 0.5)  # Xavier (gain=1, no activation)
    b2 = torch.randn(vocab_size,                       generator=g) * 0.01  # small, near zero

    # add learnable scale and shift (gain and bias) parameters for batchnorm
    # we initialize these to be 1 and 0 respectively so that at initialization of hte model, the normalized activations become a unit gaussian
    bngain = torch.ones((1, n_hidden))
    bnbias = torch.zeros((1, n_hidden))

    parameters = [C, W1, b1, W2, b2, bngain, bnbias]
    for p in parameters:
        p.requires_grad = True
    return C, W1, b1, W2, b2, bngain, bnbias, parameters

def train_model(params, n_steps=200000, batch_size=32):

    C, W1, b1, W2, b2, bngain, bnbias, parameters = params

    stepi = []
    lossi = []
    loglossi = []

    pbar = tqdm(range(n_steps), desc="Training")
    for i in pbar:
        # Mini-batch: randomly sample 32 examples
        ixds = torch.randint(0, Xtr.shape[0], (batch_size,))
        mini_batch_inp, mini_batch_target = Xtr[ixds], Ytr[ixds]

        # Forward pass
        emb = C[mini_batch_inp]                                        # (32, 3, 10)
        emb_cat = emb.view(emb.shape[0], -1)                            # (32, 30)
        hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)

        # CONVERT hidden_layer_preactivation to unit gaussian
        _mean = hidden_layer_preactivation.mean(axis=0, keepdim=True) # (1, 200) take mean across the samples in the mini batch
        _std = hidden_layer_preactivation.std(axis=0, keepdim=True) # (1, 200) take std across the samples in the mini batch

        hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std # convert to unit gaussian

        # scale and shift
        hidden_layer_preactivation = hidden_layer_preactivation*bngain + bnbias

        h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
        logits = h @ W2 + b2                                           # (32, 27)
        loss = F.cross_entropy(logits, mini_batch_target)

        # Backward pass
        for p in parameters:
            p.grad = None
        loss.backward()

        # Learning rate step decay: 0.1 for first 100K steps, then 0.01
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad

        # Track loss
        stepi.append(i)
        lossi.append(loss.item())
        loglossi.append(loss.log10().item())

        if i % 10 == 0:
            pbar.set_postfix(loss=f"{loss.data:.4f}")

    return stepi, lossi, loglossi, hidden_layer_preactivation, h, logits

C, W1, b1, W2, b2, bngain, bnbias, parameters = get_params(embed_dim, block_size, n_hidden)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits = train_model([C, W1, b1, W2, b2, bngain, bnbias, parameters])
````

### The train/eval discrepancy

We also need to update our eval code to use the learned `bngain` and `bnbias` parameters. But there's a subtler issue lurking here: during training, we compute `_mean` and `_std` from each mini-batch. This works fine when processing batches of 32 — we get a reasonable estimate of the statistics. But what happens at **inference time**?

During inference, we might want to predict a single example. Computing the mean and std of a "batch" of size 1 is meaningless — the mean would just be the value itself, and the std would be zero (division by zero!). Even with small batches, the statistics would be noisy and the predictions would change depending on what other examples happen to be in the same batch. This is undesirable — we want **deterministic** predictions that depend only on the input, not on its neighbors.


````python
@torch.no_grad()
def eval_loss(X, Y, params, embed_dim, block_size):
    C, W1, b1, W2, b2, bngain, bnbias, parameters = params

    emb = C[X]
    emb_cat = emb.view(emb.shape[0], -1)

    hidden_layer_preactivation = emb_cat @ W1 + b1
    _mean = hidden_layer_preactivation.mean(axis=0, keepdim=True)
    _std = hidden_layer_preactivation.std(axis=0, keepdim=True)
    hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std
    hidden_layer_preactivation = hidden_layer_preactivation*bngain + bnbias

    h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)
    return loss.item()

print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, bngain, bnbias, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, bngain, bnbias, parameters], embed_dim, block_size)}")
````

With the learnable scale and shift, our losses are now comparable to what we had before (Training: ~2.04, Validation: ~2.10) with only **5% saturated neurons** (down from 23%). The BatchNorm layer is doing its job — keeping the activations well-behaved while still allowing the network to learn expressive representations.

However, there is still a very weird quirk we had to introduce during evaluation. In the `eval_loss` function above, we're computing the mean and std over the _evaluation_ examples. This means the result for a single example depends on what other examples are in the batch — inference is not deterministic.

### Solution 1: Post-training calibration

The simplest fix is a two-stage approach: after training is complete, make a single pass over the entire training set to compute the "true" population mean and std for each neuron, then use those fixed statistics during inference.


````python
# stage 2 of training: calculate the batchnorm statistics

with torch.no_grad():
    emb = C[Xtr]
    emb_cat = emb.view(emb.shape[0], -1)

    hidden_layer_preactivation = emb_cat @ W1 + b1

    # measure mean and std
    bnmean = hidden_layer_preactivation.mean(axis=0, keepdim=True)
    bnstd = hidden_layer_preactivation.std(axis=0, keepdim=True)

# now we can use these during our evaluation/inference
@torch.no_grad()
def eval_loss(X, Y, params, embed_dim, block_size):
    C, W1, b1, W2, b2, bngain, bnbias, parameters = params

    emb = C[X]
    emb_cat = emb.view(emb.shape[0], -1)

    hidden_layer_preactivation = emb_cat @ W1 + b1
    hidden_layer_preactivation = (hidden_layer_preactivation - bnmean) / bnstd
    hidden_layer_preactivation = hidden_layer_preactivation*bngain + bnbias

    h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)
    return loss.item()

print(f"Final Training loss: {eval_loss(Xtr, Ytr, [C, W1, b1, W2, b2, bngain, bnbias, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [C, W1, b1, W2, b2, bngain, bnbias, parameters], embed_dim, block_size)}")
````

The results are exactly the same as before — the population statistics computed over the full training set give the same normalization as the batch-level statistics (which makes sense: the training set _is_ the population we care about).

But this approach is clunky — nobody wants to do an extra calibration pass after training. It also means we can't evaluate loss during training (e.g. for early stopping) without either using batch statistics (noisy) or running the full calibration each time (expensive).

### Solution 2: Running average (exponential moving average)

The BatchNorm paper's elegant solution: maintain a **running estimate** of the population mean and std _during training_, updated at each step via an exponential moving average (EMA):

$$\mu_\text{running} \leftarrow (1 - \alpha) \cdot \mu_\text{running} + \alpha \cdot \mu_\text{batch}$$
$$\sigma_\text{running} \leftarrow (1 - \alpha) \cdot \sigma_\text{running} + \alpha \cdot \sigma_\text{batch}$$

where $\alpha$ is a small momentum factor (we use `0.001`). The running estimates are initialized to $\mu = 0$ and $\sigma = 1$ (matching the unit Gaussian assumption), and they slowly converge to the true population statistics as training progresses.

These running estimates are **not trainable parameters** — they don't receive gradients and aren't updated by the optimizer. They're pure bookkeeping: statistics that we track for use at inference time. In PyTorch's `nn.BatchNorm1d`, these are called `running_mean` and `running_var` and are stored as **buffers** (not parameters) — they get saved with the model but aren't part of the computation graph.

At inference time, we simply use the running estimates instead of computing batch statistics, giving us deterministic predictions that work for any batch size (including single examples).


````python
# Lets add these scale and shift parameters

def get_params(embed_dim, block_size, n_hidden):

    tanh_gain = 5./3  # correction factor for tanh squashing variance

    g = torch.Generator().manual_seed(2147483647)
    C  = torch.randn((vocab_size, embed_dim),            generator=g)
    W1 = torch.randn((embed_dim * block_size, n_hidden), generator=g) * (tanh_gain / (embed_dim * block_size) ** 0.5)  # Kaiming with tanh gain
    b1 = torch.randn(n_hidden,                        generator=g) * 0.01  # small, near zero
    W2 = torch.randn((n_hidden, vocab_size),           generator=g) * (1 / (n_hidden) ** 0.5)  # Xavier (gain=1, no activation)
    b2 = torch.randn(vocab_size,                       generator=g) * 0.01  # small, near zero

    bngain = torch.ones((1, n_hidden))
    bnbias = torch.zeros((1, n_hidden))

    # add tracking variables which are NOT part of the training and thus dont receive gradients
    bnstd_running = torch.ones((1, n_hidden))
    bnmean_running = torch.zeros((1, n_hidden))

    parameters = [C, W1, b1, W2, b2, bngain, bnbias]
    for p in parameters:
        p.requires_grad = True

    return bnstd_running, bnmean_running, parameters

def train_model(params, n_steps=200000, batch_size=32):

    bnstd_running, bnmean_running, parameters = params
    C, W1, b1, W2, b2, bngain, bnbias = parameters

    stepi = []
    lossi = []
    loglossi = []

    pbar = tqdm(range(n_steps), desc="Training")
    for i in pbar:
        # Mini-batch: randomly sample 32 examples
        ixds = torch.randint(0, Xtr.shape[0], (batch_size,))
        mini_batch_inp, mini_batch_target = Xtr[ixds], Ytr[ixds]

        # Forward pass
        emb = C[mini_batch_inp]                                        # (32, 3, 10)
        emb_cat = emb.view(emb.shape[0], -1)                            # (32, 30)
        hidden_layer_preactivation = emb_cat @ W1 + b1                  # (32, 200)

        # CONVERT hidden_layer_preactivation to unit gaussian
        _mean = hidden_layer_preactivation.mean(axis=0, keepdim=True) # (1, 200) take mean across the samples in the mini batch
        _std = hidden_layer_preactivation.std(axis=0, keepdim=True) # (1, 200) take std across the samples in the mini batch

        hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std # convert to unit gaussian

        # move the running average slightly based on the currnet mean and std directions
        with torch.no_grad():
            bnmean_running = 0.999 * bnmean_running + 0.001 * _mean
            bnstd_running = 0.999 * bnstd_running + 0.001 * _std

        # scale and shift
        hidden_layer_preactivation = hidden_layer_preactivation*bngain + bnbias

        h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
        logits = h @ W2 + b2                                           # (32, 27)
        loss = F.cross_entropy(logits, mini_batch_target)

        # Backward pass
        for p in parameters:
            p.grad = None
        loss.backward()

        # Learning rate step decay: 0.1 for first 100K steps, then 0.01
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad

        # Track loss
        stepi.append(i)
        lossi.append(loss.item())
        loglossi.append(loss.log10().item())

        if i % 10 == 0:
            pbar.set_postfix(loss=f"{loss.data:.4f}")

    return stepi, lossi, loglossi, hidden_layer_preactivation, h, logits, bnmean_running, bnstd_running

bnstd_running, bnmean_running, parameters = get_params(embed_dim, block_size, n_hidden)

stepi, lossi, loglossi, hidden_layer_preactivation, h, logits, bnmean_running, bnstd_running = \
    train_model([bnstd_running, bnmean_running, parameters])
````

````python
# now we can use these running mean and std during eval
@torch.no_grad()
def eval_loss(X, Y, params, embed_dim, block_size):

    bnstd_running, bnmean_running, parameters = params
    C, W1, b1, W2, b2, bngain, bnbias = parameters

    emb = C[X]
    emb_cat = emb.view(emb.shape[0], -1)

    hidden_layer_preactivation = emb_cat @ W1 + b1
    hidden_layer_preactivation = (hidden_layer_preactivation - bnmean_running) / bnstd_running
    hidden_layer_preactivation = hidden_layer_preactivation*bngain + bnbias

    h = torch.tanh(hidden_layer_preactivation)  # (32, 200)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)
    return loss.item()

print(f"Final Training loss: {eval_loss(Xtr, Ytr,[bnstd_running, bnmean_running, parameters], embed_dim, block_size)}, \
    Validation Loss: {eval_loss(Xdev, Ydev, [bnstd_running, bnmean_running, parameters], embed_dim, block_size)}")
````

We can see that the losses are almost exactly the same as before. There can be a slight variation since we are calculating a running average as opposed to true average.

## Summary: The full BatchNorm recipe

We built up Batch Normalization in three steps, each solving a problem introduced by the previous:

| Step                             | What we did                                                                | Problem it solved                                   | Problem it introduced                     |
| -------------------------------- | -------------------------------------------------------------------------- | --------------------------------------------------- | ----------------------------------------- |
| **1. Raw normalization**         | Normalize pre-activations to $\mu=0, \sigma=1$ per neuron across the batch | Eliminated neuron saturation (<1% vs 23%)           | Over-constrained the network → worse loss |
| **2. Learnable $\gamma, \beta$** | Add scale and shift parameters after normalization                         | Restored expressivity → loss back to ~2.04          | Inference depends on batch composition    |
| **3. Running EMA**               | Track $\mu_\text{running}, \sigma_\text{running}$ during training          | Deterministic inference, no calibration step needed | —                                         |

The complete BatchNorm forward pass for a single layer is:

$$\hat{x} = \frac{x - \mu_B}{\sigma_B} \quad \text{(normalize)} \qquad y = \gamma \cdot \hat{x} + \beta \quad \text{(scale and shift)}$$

where $\mu_B, \sigma_B$ are batch statistics during training and running estimates during inference.


### A note on `b1` with BatchNorm

Lets look at the normalization through the neuron

```python
    hidden_layer_preactivation = emb_cat @ W1 + b1
    _mean = hidden_layer_preactivation.mean(axis=0, keepdim=True)
    _std = hidden_layer_preactivation.std(axis=0, keepdim=True)
    hidden_layer_preactivation = (hidden_layer_preactivation - _mean) / _std
```

One subtle detail: when BatchNorm is applied after a linear layer (`emb_cat @ W1 + b1`), the bias `b1` becomes **redundant**. The normalization step subtracts the mean, which cancels out any constant offset added by `b1`. The learnable $\beta$ in BatchNorm then takes over the role of the bias. In practice, when using BatchNorm, you should set `bias=False` in the preceding linear layer (e.g., `nn.Linear(30, 200, bias=False)`) to avoid wasting parameters.

We can confirm this by looking at the gradient of `b1` - it should be 0.

### Looking ahead

BatchNorm was a landmark innovation, but it has known limitations:

- **Batch size dependence:** Small batches give noisy statistics, hurting performance. This is a problem for tasks where memory limits batch size (e.g., object detection with large images).
- **Train/eval mode mismatch:** The running statistics can diverge from the true statistics if the training distribution shifts, leading to subtle bugs.
- **Sequential models:** For RNNs and Transformers, the "batch" dimension doesn't align well with the temporal structure.

These limitations motivated alternatives like **Layer Normalization** (used in Transformers), **Group Normalization** (used in vision when batch sizes are small), and **RMSNorm** (a simplified variant used in LLaMA and other modern LLMs). We'll encounter some of these in later notebooks.
